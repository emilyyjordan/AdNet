```python
import random
import math
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
random.seed(10)
myT = []
#for i in range(100):
    #t = np.random.randint(0,2) #1 or 0 to represent positive or negative values
    #myT.append(t)


count1 = 1
count2 = 1
while count1 <= 50 or count2 <= 50:
    choice = np.random.randint(0,2)
    print(choice)
    if choice == 0 and count1 <= 50:
        count1 += 1
        myT.append(0)
        print("while loop 1")

    if choice == 1 and count2 <= 50:
        count2 += 1
        myT.append(1)
        print("while loop 2")
```

```python
#deck A - addictive
wholeDeckA = []
for i in range(1, 101):
    t = myT[i-1]
    err = np.random.normal(0,5)
    if t == 1:
        currCard = -20 + err
        wholeDeckA.append(currCard)
    else:
        currCard =(2000/(math.log((i+5), 1.1)))-35
        wholeDeckA.append(currCard)

wholeDeckA = [(5 * round(i/5)) for i in wholeDeckA]

    

#plt.tight_layout(sns.despine()
#plt.subplot(1,3,1)
#plt.tight_layout()
plt.scatter(range(len(wholeDeckA)), wholeDeckA, color = "#3778bf")
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.show()

#deck B - non-addictive
wholeDeckB = []
for i in range(100):
    t = myT[i]
    err = np.random.normal(0, 5)
    if t == 1:
        currCard = 50 + err #move 50 closer to 0
        wholeDeckB.append(currCard)
    else:
        currCard =(-2000/(math.log((i+5), 1.1)))+35
        wholeDeckB.append(currCard)
        
wholeDeckB = [(5 * round(i/5)) for i in wholeDeckB]


#plt.tight_layout()
#sns.despine()
#plt.subplot(1,3,2)
#plt.tight_layout()
#plt.scatter(range(len(wholeDeckB)), wholeDeckB, color = "#9b59b6")#purple
plt.scatter(range(len(wholeDeckB)), wholeDeckB, color = "#feb308") #yellow
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.show()

#deck C - control
wholeDeckC = []
for i in range(100):
    t = myT[i]
    beta = np.random.normal(0, 3)
    if t == 1:
        if i%2 == 0:
            currCard =(30+beta)/(t)
    
        else:
            currCard =(30-beta)/(t)
        wholeDeckC.append(currCard)
    else:
        if i%2 == 0:
            currCard =(-30+beta)

        else:
            currCard =(-30-beta)
        wholeDeckC.append(currCard)

wholeDeckC = [(5 * round(i/5)) for i in wholeDeckC]


#plt.tight_layout()
#sns.despine()
#plt.subplot(1,3,3)
#plt.tight_layout()
plt.scatter(range(len(wholeDeckC)), wholeDeckC, color = "#3778bf")
plt.scatter(range(len(wholeDeckC)), wholeDeckC, color = "#9b59b6")
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.show()

#plt.figure(figsize = (len(wholeDeckC), wholeDeckC)) 

```

```python
df1 = pd.DataFrame()
df1["Deck A"] = wholeDeckA
df1["Deck B"] = wholeDeckB
df1["Deck C"] = wholeDeckC
df1.to_excel("deckResults1.xlsx", index = False)
print(sum(wholeDeckA))
print(sum(wholeDeckB))
print(sum(wholeDeckC))
```
